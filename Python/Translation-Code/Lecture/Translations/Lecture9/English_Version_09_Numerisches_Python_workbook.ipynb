#### 9.2.1 Creating arrays with `arange()`

The syntax of `arange()` is:

```python
arange([start=0,] stop[, step=1][, dtype = None])
```

where:

- *evenly distributed* values ​​are generated within the half-open interval `[start, stop)`
- Both `int` and `float` values ​​can be passed as arguments
- the type of the array can be determined with `dtype`. If nothing is specified, the type is determined based on the input values

If `dtype = int`, the `arange()` function is almost equivalent to the built-in `range()` function. The difference lies in the return object. `arange()` returns an `ndarry`, while `range()` returns a `range` object, which is an *iterator*.

````python
# TODO example to arange()


# in comparison range():
x = range(1,7)
print(x)
print(list(x))

# TODO more arange examples


# strange examples
print("\nSeltsame Beispiel")
b = np.arange(12.04, 12.84, 0.08)
print("b, ",b)
c = np.arange(0.6, 10.4, 0.71, int)
print("c: ",c)
d = np.arange(0.28, 5, 0.71, int)
print("d: ",d)
#help(np.arange)
````

#### 9.2.2 Creating arrays with `linspace()`

The syntax of `linspace()` is:

```python
linspace(start, stop[, num=50][, endpoint = True][, retstep=False])
```

where:

- `endpoint` specifies whether the value `stop` is also output or not. Default = `True`
- `num` specifies the number of *evenly distributed* values ​​from the interval `[start, stop]`. Default = `50`
- for `retstep=True` the distance between the returned values ​​of the array is also returned. Default = `False`

````python
# TODO examples for linspace()
# optional for “beautification”
np.set_printoptions(linewidth=65, precision=3)
print("\nLinspace Beispiel")
print(np.linspace(1, 10, 10))
print(np.linspace(1, 10, 10, False))
print(np.linspace(1, 10, 10, True))
print(np.linspace(1, 10, 10, False, True))
print(np.linspace(1, 10, 10, True, True))
print(np.linspace(1, 10, 10, False, False))
print(np.linspace(1, 10, 10, True, False))


````

Why is the `retstep` parameter needed anyway? Can't you just calculate the difference between two adjacent values?

````python
# Example why retstep makes sense
a, spacing = np.linspace(4, 23, endpoint=False, retstep=True)
print(a[:6])
# Distances between the first 6 array elements:
for i in range(6):
    print(a[i + 1] - a[i])
print(f"{spacing=}")
````

#### 9.2.4 Changing the shape of an array

`numpy` arrays have methods that construct a *view* with a new shape: `reshape()` and `ravel()`. The new form is not its own independent object, but rather a *view* that remains connected to the original array. When the original array is changed, the *view* also changes.

- When calling `reshape()`, natural numbers are passed that describe the shape of the desired view

- `ravel()` returns a *flattened* view, a one-dimensional array in which all elements of the original array are written one after the other.

````python
# TODO Example of reshape() and ravel()
a = np.arange(6)
print(a)

b= a.reshape(3,2)
print(b)

c =b.ravel()
print(c)
````

The `resize()` method is used to change the array (and its size). The arguments are numbers that describe the desired form. If the new shape is larger, zeros are added. However, it is recommended to call the method via `np.resize(array, (newsize))` instead of directly calling it via `ndarray` via `ndarray.resize((newsize))`. Otherwise, especially with arrays created via `linspace()`, this can lead to error messages. This involves filling up with repeated copies of the array until the desired size is reached (see also [Documentation](https://numpy.org/doc/stable/reference/generated/numpy.resize.html#numpy-resize) )

````python
# Change TODO array
x.shape = (6, 3)
print(x)

````

#### 9.2.5 Indexing and slicing

To access individual elements, use the `[]` operator as with *sequences*:

````python
# Access elements of an array
f = np.array([1, 7, 2, 3, 5, 8, 13, 21])
print(f[0]) # erstes Element
print(f[1]) # zweites Element
print(f[-1])# letztes Element
````

````python
# Index multidimensional arrays
y = np.array([[[111, 112], [121, 122]], 
              [[211, 212], [221, 222]], 
              [[311, 312], [321, 322]]]) 

print(y[0][1][0])   # wie bei verschachtelten Listen
print(y[0, 1, 0])   # funktioniert genauso
````

The *slicing* also works similarly to normal Python sequences according to the syntax `[start:stop:step]`. **Attention**: While new objects are created with lists and tuples, the subrange operator in NumPy only generates a *view* of the original array. To copy an array you need the `copy()` method.

````python
# Some examples
a = np.arange(9)
print(a)
print(a[1:4])
print(a[:4])
print(a[4:])
print(a[::2])
b = a[:]    # Sicht auf a
print(b)
b[0] = 99
print(a)

# Copy TODO (partial) array
````

#### 9.2.7 Concatenation of arrays

Syntax der Funktion `concatenate`:

```python
concatenate((a1, a2, ...), axis=0,
            out=None, dtype=None)
```

````python
# Example of concatenation
# one-dimensional arrays
x = np.array([11, 22])
y = np.array([18, 7, 6])
z = np.array([1, 3, 5])
c = np.concatenate((x, y, z))
print(c)
````

````python
# mehrdimensionale Arrays, axis=0
x = np.array(range(8))
x = x.reshape((2, 4))
y = np.array(range(100, 112))
y = y.reshape((3, 4))
z = np.concatenate((x, y)) # axis=1 geht nicht
print(f'x:\n{x}\n\ny:\n{y}\n\nz:\n{z}')
````

````python
# mehrdimensionale Arrays, axis=1
x = np.array(range(8)).reshape((4, 2))
y = np.array(range(100, 112)).reshape((4, 3))
z = np.concatenate((x, y), axis=1)
print(f'x:\n{x}\n\ny:\n{y}\n\nz:\n{z}')
````

? When concatenating multidimensional arrays along a specific axis, it must be ensured that the *shape* of **all other dimensions** are the same.

````python
# three-dimensional example
x = np.array(range(24)).reshape((2, 3, 4))
y = np.array(range(100, 116)).reshape((2, 2, 4))
z = np.concatenate((x, y), axis=1)
print(f'x:\n{x}\n\ny:\n{y}\n\nz:\n{z}')
````

#### 9.2.8 Comparisons

Two arrays can be compared element by element using the usual comparison operators (`<, <=, >, >=, ==, !=`). The result is an array with truth values:

````python
# Compare TODO arrays
a = np.array([[1, 2], [2, 0]])
b = np.array([[1, 2], [2, 3]])
````

You can construct compound conditions using the `logical_or()` and `logical_and()` methods. The normal logical operators like `&` or `|` cannot be used:

````python
# Link conditions
a = np.arange(8)
b = np.logical_and(a >= 3, a <= 5)
print(b)
````
